## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# - Fill missing 'director', 'cast', 'country' with 'Unknown'
# - Drop rows with missing 'type', 'title', 'date_added', or 'rating' if necessary
data['director'].fillna('Unknown', inplace=True)
data['cast'].fillna('Unknown', inplace=True)
data['country'].fillna('Unknown', inplace=True)
data['rating'].fillna('Not Rated', inplace=True)

# Convert to datetime safely
data['date_added'] = pd.to_datetime(data['date_added'].astype(str).str.strip(), errors='coerce')

# Extract year and month from valid dates
data['added_year'] = data['date_added'].dt.year
data['added_month'] = data['date_added'].dt.month


# Clean Duration Field

# Separate 'duration' into numeric and unit (e.g., "90 min", "1 Season")
data[['duration_int', 'duration_type']] = data['duration'].str.extract('(\d+)\s*(\w+)')
data['duration_int'] = pd.to_numeric(data['duration_int'])


# Standardize Categorical Values
# Strip whitespace and make lowercase for consistency
data['type'] = data['type'].str.strip().str.title()
data['rating'] = data['rating'].str.strip().str.upper()
data['country'] = data['country'].str.strip()
data['listed_in'] = data['listed_in'].str.strip()

# Drop rows where 'date_added' is missing (and hence added_year/month too)
data.dropna(subset=['date_added'], inplace=True)


# Final Checks

print("\nFinal Shape:", data.shape)
print("\nRemaining Nulls:\n", data.isnull().sum())

# Save cleaned data if needed
data.to_csv("netflix_cleaned.csv", index=False)
````

#### Chart - 7

````python
# Chart - 7 visualization code
# Top 10 Genres by Content Type (Movie vs TV Show)

# Step 1: Explode genre into individual rows
genre_expanded = data.copy()
genre_expanded['listed_in'] = genre_expanded['listed_in'].str.split(', ')
genre_expanded = genre_expanded.explode('listed_in')

# Step 2: Group by genre and type
genre_type_counts = genre_expanded.groupby(['listed_in', 'type']).size().reset_index(name='count')

# Step 3: Select top 10 genres overall
top_genres = genre_expanded['listed_in'].value_counts().head(10).index
genre_type_counts = genre_type_counts[genre_type_counts['listed_in'].isin(top_genres)]

# Step 4: Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=genre_type_counts, x='listed_in', y='count', hue='type', palette='coolwarm')

# Customize plot
plt.title('Top 10 Genres by Content Type (Movies vs TV Shows)', fontsize=14, fontweight='bold')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.legend(title='Content Type')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
````
